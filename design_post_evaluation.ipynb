```python
def lorentzian( x, x0,a,alpha):
    return a * alpha**2 / ((alpha)**2 +(x-x0)**2)

from scipy.optimize import curve_fit

x_array = neff_array_lipson
y_array = np.array(FOM_list_lipson)

a = 1
x0 =  3.818
alpha = 1E-4

popt_lorentz, pcov_lorentz = curve_fit(lorentzian, x_array, y_array, p0=[x0, a, alpha])
perr_lorentz = np.sqrt(np.diag(pcov_lorentz))
```

```python
print("Estimated amplitude: "+str(popt_lorentz[1])+" (+/-) "+ str(perr_lorentz[1]))
print("Estimated center wavenumber: "+str(popt_lorentz[0])+" (+/-) "+ str(perr_lorentz[0]))
print("Estimated losses: "+str(popt_lorentz[2])+" (+/-) "+ str(perr_lorentz[2]))
```

```python
fig, ax = plt.subplots(figsize=(14,10))

ax.scatter(x_array, y_array, alpha=0.5, c="red", label="Simulation data")

x = np.linspace(deltamin, deltamax, 1000)

ax.plot(x, lorentzian(x,popt_lorentz[0], popt_lorentz[1] ,popt_lorentz[2]), linewidth=2, color="blue", label="Lorentzian fit")
ax.set_yscale("log")
ax.legend(frameon=True)
fig
```
