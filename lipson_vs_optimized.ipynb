```python
import numpy as np
from solver import freq_top_opt_2D
import matplotlib.pyplot as plt
%matplotlib inline
verbose = False
if not verbose:
    import warnings
    warnings.filterwarnings("ignore")
```

```python
L = 200E-6
power_array = np.array([1.0,2.5,5,10,20,40,80,160, 320])*1E-3
p_L = power_array / L
resonance_position_4 = np.array([3.8194, 3.8195, 3.8197, 3.8200, 3.8208, 3.8223, 3.8253, 3.8314, 3.8438]) -3.8193
```

```python
plt.rcParams.update(plt.rcParamsDefault)
plt.style.use("science")
import matplotlib as mpl
mpl.rcParams.update({"font.size": 22})

fig, ax = plt.subplots(figsize=(10,6))

ax.set_ylabel("$\\Delta n_\\text{eff,T}$")
ax.set_xlabel("$P_\\text{in}/L$")

ax.scatter(p_L, resonance_position_4, color="black", label="$4\,\\mu m$ metal")
ax.plot(p_L, resonance_position_4, color="black", linestyle="dashed")
ax.legend(frameon=True)
ax.set_xscale("log")
ax.set_yscale("log")


fig
```
